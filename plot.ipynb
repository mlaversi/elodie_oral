```python
import numpy as np
from scipy.integrate import odeint
```

```python

# Définition des équations différentielles
def equations(y, t, alpha, beta, gamma, N_max, B_max):
    N, B = y
    dNdt = alpha * B * (1 - N / N_max) - beta * N
    dBdt = gamma * B * (1 - B / B_max)
    return [dNdt, dBdt]

# Paramètres
alpha = 0.01  # Coefficient de croissance des pommes en fonction de la croissance des branches
beta = 0.005  # Coefficient de perte de pommes
gamma = 0.1   # Taux de croissance des branches
N_max = 1000  # Capacité maximale de production de pommes
B_max = 500   # Capacité maximale de croissance des branches

# Conditions initiales
N0 = 10  # Nombre initial de pommes
B0 = 5   # Croissance initiale des branches

# Temps
t = np.linspace(0, 60, 600)  # Durée de simulation : 60 semaines, soit 600 unités de temps

# Résolution des équations différentielles
y0 = [N0, B0]
solution = odeint(equations, y0, t, args=(alpha, beta, gamma, N_max, B_max))
N, B = solution[:, 0], solution[:, 1]

# Affichage des résultats
print("Nombre de pommes (N) : ", N)
print("Croissance des branches (B) : ", B)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Définition des équations différentielles
def equations(y, t, alpha, beta, gamma, B_max):
    N, B = y
    dNdt = alpha * B - beta * N
    dBdt = gamma * B * (1 - B / B_max)
    return [dNdt, dBdt]

# Paramètres
alpha = 0.01  # Coefficient de croissance des pommes en fonction de la croissance des branches
beta = 0.005  # Coefficient de perte de pommes
gamma = 0.1   # Taux de croissance des branches
B_max = 500   # Capacité maximale de croissance des branches

# Conditions initiales
N0 = 0  # Nombre initial de pommes
B0 = 15   # Croissance initiale des branches

# Temps
t = np.linspace(0, 200, 200)  # Durée de simulation : 3 semaines

# Résolution des équations différentielles
y0 = [N0, B0]
solution = odeint(equations, y0, t, args=(alpha, beta, gamma, B_max))
N, B = solution[:, 0], solution[:, 1]

# Affichage des résultats
plt.figure(figsize=(10, 6))
plt.plot(t, N, label='Nombre de pommes')
plt.plot(t, B, label='Croissance des branches')
plt.xlabel('Temps (jours)')
plt.ylabel('Nombre')
plt.title('Évolution du nombre de pommes et de la croissance des branches sur 3 semaines')
plt.legend()
plt.grid(True)
plt.show()

# Nombre de pommes et de branches après 3 semaines
print("Nombre de pommes après 3 semaines:", N[-1])
print("Croissance des branches après 3 semaines:", B[-1])
```
